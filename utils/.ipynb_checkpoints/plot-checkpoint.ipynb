```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fedmach_prec_list=np.load('../save/fedmach_Eurlex_500_4.npy')
fedavg_prec_list=np.load('../save/fedavg_Eurlex.npy')
plt.figure()
plt.plot(range(len(fedmach_prec_list)), fedmach_prec_list[:,0], 'r-', label='FedHash: Precision@1')
plt.plot(range(len(fedmach_prec_list)), fedmach_prec_list[:,1], 'y-', label='FedHash: Precision@3')
plt.plot(range(len(fedmach_prec_list)), fedmach_prec_list[:,2], 'b-', label='FedHash: Precision@5')
plt.plot(range(len(fedavg_prec_list)), fedavg_prec_list[:,0], 'r--', label='FedAvg: Precision@1')
plt.plot(range(len(fedavg_prec_list)), fedavg_prec_list[:,1], 'y--', label='FedAvg: Precision@3')
plt.plot(range(len(fedavg_prec_list)), fedavg_prec_list[:,2], 'b--', label='FedAvg: Precision@5')
plt.title('Non-iid Data Partition')
plt.legend()
plt.ylabel('Precision')
plt.savefig('../save/fedmach_vs_fedavg_epc_100_B500_R4_test_acc_noniid.png')
```
